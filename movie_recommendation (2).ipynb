```python
import pandas as pd

movies = pd.read_csv(r"C:\Users\shwet\Downloads\ml-latest-small (1)\ml-latest-small\movies.csv")
ratings = pd.read_csv(r"C:\Users\shwet\Downloads\ml-latest-small (1)\ml-latest-small\ratings.csv")

# Merge data
data = pd.merge(ratings, movies, on='movieId')

# Create matrix
movie_matrix = data.pivot_table(index='userId', columns='title', values='rating')
movie_matrix = movie_matrix.fillna(0)

movie_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(movie_matrix.T)
similarity_df = pd.DataFrame(similarity, index=movie_matrix.columns, columns=movie_matrix.columns)

similarity_df.head()
```

```python
def recommend(movie_name):
    return similarity_df[movie_name].sort_values(ascending=False).head(10)
```

```python
recommend("Toy Story (1995)")
```

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split

# Load dataset
reader = Reader(rating_scale=(0.5, 5))
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)

# Split data
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
import numpy as np

# Create matrix
matrix = movie_matrix.values

# Apply SVD
U, sigma, Vt = np.linalg.svd(matrix, full_matrices=False)

# Convert sigma to diagonal matrix
sigma = np.diag(sigma)

# Reconstruct matrix
predicted_ratings = np.dot(np.dot(U, sigma), Vt)

pred_df = pd.DataFrame(predicted_ratings, columns=movie_matrix.columns)

pred_df.head()
```

```python
def recommend_svd_simple(movie_name):
    return pred_df[movie_name].sort_values(ascending=False).head(10)
```

```python
recommend_svd_simple("Toy Story (1995)")
```

```python
def recommend_svd_simple(movie_name):
    similar_scores = pred_df[movie_name].sort_values(ascending=False).head(10)
    
    # convert index to movie names
    return similar_scores.reset_index().rename(columns={'index': 'Movie', movie_name: 'Score'})
```
